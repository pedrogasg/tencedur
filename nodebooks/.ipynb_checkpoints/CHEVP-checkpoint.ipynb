```python
import cv2
import numpy as np
import scipy.interpolate as si
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('road.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
#canny basic 60,150,3
blur = cv2.medianBlur(gray,5)
edges = cv2.Canny(blur,60,150,3)
plt.imshow(edges, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
rows, cols = edges.shape
u,d,ul,ur,dl,dr = 0.5, 0.01, 0.4, 0.4, -.5, -.5
bottom_left  = [cols*dl, rows*(1.-d)]
bottom_right = [cols*(1-dr), rows*(1.-d)]
top_left     = [cols*ul, rows*u]
top_right    = [cols*(1-ur), rows*u] 
vertices = np.array([[bottom_left, top_left, top_right, bottom_right]], dtype=np.int32)
mask = np.zeros_like(edges)
cv2.fillPoly(mask, vertices, 255)
work = cv2.bitwise_and(edges, mask)
```

```python
def get_segments(img, sizes):
    mark = 0
    segments = []
    for size in segments_size:
        print('from:'+str(mark)+' to:'+str(mark+size))
        segments.append(img[mark:mark+size, :])
        mark += size
    return segments
```

```python
height,width = edges.shape
proportions = [0.50, 0.10, 0.125, 0.125, 0.15]
segments_size = [int(p * height) for p in proportions] 

segments = get_segments(work, segments_size)
```

```python
def get_vector(xs,ys,xe,ye):
    k = (ye - ys)/(xe - xs)
    b = ys - k * xs
    return (k, b)

def horizontal_intersection(xs, ys, xe, ye, x):
    k, b = get_vector(xs,ys,xe,ye)
    if k == 0 :
        return None
    y = k*x+b
    if y > ys and y > ye :
        return None
    return (x, int(y))

def find_intersection(l1, l2):
    xs1,ys1,xe1,ye1 = l1
    xs2,ys2,xe2,ye2 = l2
    
#   s1_x = xe1 - xs1
#   s1_y = ye1 - ys1
    
#   s2_x = xe2 - xs2
#   s2_y = ye2 - ys2
    
#   s = (-s1_y * (xs1 - xs2) + s1_x * (ys1 - ys2)) / (-s2_x * s1_y + s1_x * s2_y)
#   t = ( s2_x * (ys1 - ys2) - s2_y * (xs1 - xs2)) / (-s2_x * s1_y + s1_x * s2_y)

    h1 = xs1 - xe1 == 0
    h2 = xs2 - xe2 == 0
    
    if h1 and h2:
        return None
    
    if h1:
        return horizontal_intersection(xs2,ys2,xe2,ye2,xs1)
    
    if h2:
        return horizontal_intersection(xs1,ys1,xe1,ye1,xs2)
    
    k1, b1 = get_vector(xs1,ys1,xe1,ye1)
    k2, b2 = get_vector(xs2,ys2,xe2,ye2)
    
    if (k1 == k2 or k1 == 0 or k2 == 0):
        return None

    xi = (b2-b1)/(k1-k2)
    yi = k1*xi+b1
    
    return int(xi), int(yi)#, int(s), int(t)
```

```python
horizons = np.zeros((height*2,1))
origin_height = 0 
hough_threshold = [20, 30, 30, 40, 40]
hough_minLineLength = [10, 10, 10, 10, 10]
hough_maxLineGape = [6,6,6,6,6]
raw_lines = []
skipSegments = []
group_size=12
padding = 5
angle_threshold = 5
draw = True
for index , (size, segment, thresh, minL, maxL) in enumerate(zip(segments_size, segments, hough_threshold,hough_minLineLength, hough_maxLineGape)) : 
    if (index in skipSegments) :
        origin_height += size; raw_lines.append(None)
        continue
    lines_segment = []
    lines = cv2.HoughLinesP(segment,1,np.pi/180,thresh,minL,maxL)
    if lines is not None:
        for line in lines:
                for x1,y1,x2,y2 in line:
                    angle = np.arctan2(y2 - y1, x2 - x1) * 180. / np.pi
                    if(width-padding > x1 or width-padding > x2) and (padding < x1 or padding < x2) and (angle > angle_threshold or angle < -angle_threshold):
                        lines_segment.append(line)

    raw_lines.append(lines_segment)

    if lines_segment is not None : # it can happen that there are no lines in a section
        length = len(lines_segment)
        for i in range(length) :
            for j in range(i+1,length) :
                intersection_point = find_intersection(lines_segment[i][0],lines_segment[j][0])
                if intersection_point is None:
                    continue
                x,y = intersection_point
                vanish_point = origin_height + y
                if vanish_point > -height and vanish_point < height :
                    horizons[height+vanish_point] += 1

        # optional -- draw all lines 
        if (draw) :
            for line in lines_segment :
                for x1,y1,x2,y2 in line :
                    cv2.line(segment,(x1,y1),(x2,y2),(255,255,255),5)
    origin_height += size

best_group = np.argmax(np.sum(horizons.reshape(-1, group_size), axis=1))
horizon = best_group*group_size - height

```

```python
if horizon < 0:
    h = -horizon + 20
    work_horizon = np.concatenate((np.zeros((h,width)), work.copy()), 0)
    hz = 20
else:
    hz = horizon 
    work_horizon = work.copy()
cv2.line(work_horizon, (0,hz), (width,hz), (255,255,255),5)
plt.imshow(work_horizon, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
def right_size(line, width) :
    x1,y1,x2,y2 = line
    return (x1+x2)/2 >= width*0.46

def most_right(line1, line2) :
    return min(line1[0],line1[2]) < min(line2[0],line2[2])

def most_left(line1, line2) :
    return max(line1[0],line1[2]) > max(line2[0],line2[2])
        
def pick_pair(left, right, old_left, old_right, width) :
    if old_left is None :
        return left, right
    elif most_right(right, old_right) \
            and most_left(left, old_left) :
        return left, right
    else :
        return old_left, old_right
```

```python
origin_height = height
best_right = []
best_left = []
for lines_segment,size,segment in zip(reversed(raw_lines),reversed(segments_size),reversed(segments)):
    origin_height -= size
    left = None
    right = None

    if lines_segment is None :
        best_right.append(right)
        best_left.append(left)
        continue

    righty = [x[0] for x in lines_segment if right_size(x[0],width)]
    lefty = [x[0] for x in lines_segment if not right_size(x[0],width)]
    length = len(lines_segment)
    for l in lefty :
        for r in righty :
            intersection_point = find_intersection(l,r)
            if intersection_point is None:
                continue
            x,y = intersection_point
            vanish_point = origin_height + y
            if horizon <= vanish_point and vanish_point <= horizon+group_size :
                left, right = pick_pair(l,r,left,right,width)
    best_right.append(right)
    best_left.append(left)

print(best_right) 
print(best_left)
```

```python
best = np.zeros_like(edges)
segments2 = get_segments(best, segments_size)
for s, r, l in zip(reversed(segments2), best_right, best_left):
    if r is None or l is None:
        continue

    cv2.line(s,(r[0],r[1]),(r[2],r[3]),(255,255,255),5)
    cv2.line(s,(l[0],l[1]),(l[2],l[3]),(255,255,255),5)

plt.imshow(best, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
def get_x(line, y, zero) :
    if line[0] == line[2] :
        return line[0]
    else :
        xs,ys,xe,ye = left_line
        k,b = get_vector(xs,ys,xe,ye)
        if k == 0 :
            return zero
        else :
            return (y-b)/k

def find_midpoint(left_line, right_line, origin_height, size, horizon, width, sk, nonempty_sections) :

    if origin_height < horizon :
        origin_height = horizon
    y = size
    xl = get_x(left_line, size, 0)
    xr = get_x(right_line, size, width)

    if size != 0 and nonempty_sections == 0:
        sk = (xr-xl)/(y+origin_height-horizon)
        return int(0.5*(xr-xl) + xl), size + origin_height, sk, nonempty_sections
    
    nonempty_sections += 1
    return int(0.5*(xr-xl) + xl), y + origin_height, sk, nonempty_sections
```

```python
p = None
origin_height = height
flag = True
nonempty_sections = 0
k = None
xm = []
ym = []
n = len(segments_size)
for left_line,right_line, size,section in zip(best_left, best_right, reversed(segments_size),reversed(segments)) :
    origin_height -= size

    if left_line is None or right_line is None :
        xm.append(None)
        ym.append(None)
        continue
        
    if nonempty_sections == 0 :
        x,y,k,nonempty_sections = find_midpoint(left_line,right_line,origin_height, size, horizon,width,k,nonempty_sections)
        xm.append(x)
        ym.append(y)

    x,y,k,nonempty_sections = find_midpoint(left_line,right_line,origin_height, 0, horizon,width,k,nonempty_sections)
    xm.append(x)
    ym.append(y)

# find sections without lines
origin_height = height-segments_size[n-1]
flag = 0
last_found = 0
for j, size in zip(range(1,n), reversed(segments_size[:n-1])) :
    if j+flag < len(xm) and xm[j+flag] is not None and xm[j+flag-1] is None:
        i = j-1
        current = origin_height
        left_line = best_left[i+1]
        right_line = best_right[i+1] 
        for k,size in zip(reversed(range(last_found,i+1)),reversed(segments_size[n-i-1:n-last_found])) :
            x,y,k,nonempty_sections = find_midpoint(left_line,right_line,current, size, horizon,width,k,nonempty_sections)
            xm[k+flag] = x; ym[k+flag] = y
            current += size

        if flag == 0: 
            flag = 1
        last_found = j
    elif j+flag < len(xm) and xm[j+flag] is not None :
        last_found = j+1
    origin_height -= size

while len(xm) and xm[-1] is None :
    xm.pop()
    ym.pop()
while len(xm) and ym[-1] == ym[-2] and ym[-1] == horizon :
    xm.pop()
    ym.pop()
print(k)    
print(xm)
print(ym)
```

```python
def bspline(cv, n=200, degree=3, periodic=False):
    cv = np.asarray(cv)
    count = len(cv)

    if periodic:
        factor, fraction = divmod(count+degree+1, count)
        cv = np.concatenate((cv,) * factor + (cv[:fraction],))
        count = len(cv)
        degree = np.clip(degree,1,degree)

    # If opened, prevent degree from exceeding count-1
    else:
        degree = np.clip(degree,1,count-1)


    # Calculate knot vector
    kv = None
    if periodic:
        kv = np.arange(0-degree,count+degree+degree-1,dtype='int')
    else:
        kv = np.array([0]*degree + list(range(count-degree+1)) + [count-degree]*degree,dtype='int')

    # Calculate query range
    u = np.linspace(periodic,(count-degree),n)


    # Calculate result
    arange = np.arange(len(u))
    points = np.zeros((len(u),cv.shape[1]))
    for i in range(cv.shape[1]):
        points[arange,i] = si.splev(u, (kv,cv[:,i],degree))

    return points
```

```python
def get_point(xm, ym, horizon, width):
    x1, y1 = xm[-2], ym[-2]
    x2, y2 = xm[-3], ym[-3]
    vp1 = find_intersection([xm[-1], ym[-1],x1,y1], [0,horizon, width, horizon])
    vp2 = find_intersection([x1,y1,x2,y2], [0,horizon, width, horizon])
    vp3 = find_intersection([xm[-4], ym[-4],x2,y2], [0,horizon, width, horizon])

    if vp1 is None and vp2 is None :
        return x2, y2
    elif vp2 is None and vp3 is None :
        return x1, y1
    else :
        return int((x1+x2)/2.), int((y1+y2)/2.)

def get_coordinate(p, m):
    return 1.5*p - 0.25*(m[0] + m[-1])

def get_control_point(px, py, xm, ym):
    qx = get_coordinate(px, xm)
    qy = get_coordinate(py, ym)

    Qx = np.array([xm[-1],xm[-1],xm[-1], qx,xm[0],xm[0],xm[0]]).reshape((7,1))
    Qy = np.array([ym[-1],ym[-1],ym[-1], qy,ym[0],ym[0],ym[0]]).reshape((7,1))
    midline = np.rint(bspline(np.column_stack((Qx,Qy)))).astype(int)
    return Qx, Qy, midline
```

```python
if len(xm) != 0 :
    px, py = get_point(xm, ym, horizon, width) if len(xm) > 3 else (xm[-2], ym[-2])
    Qx, Qy, midline = get_control_point(px, py, xm, ym)
```

```python
if (len(midline) != 0) :
    p = np.asarray(edges).astype('int8')
    dy, dx = np.gradient(p)
    gamma = 0.2e-3
    s_max = float(len(midline))
    n = 2
    m = 5
    C = np.array([[-1/6.,0.5,-0.5,1/6.],[0.5,-1,0.5,0],[-0.5,0,0.5,0],[1/6.,2/3.,1/6.,0]])
    done = False
    kk = 0
    while(not done) :
        done = True
        kk += 1
        ek = 0
        sQx = np.zeros(Qx.shape)
        sQy = np.zeros(Qx.shape)
        for i in range (-1, 3) :
            ii = i+1
            mi = np.zeros((m,4))
            eix = np.zeros((m,1)); eiy = np.zeros((m,1))
            for j in range(m) :
                init = ii*40 + 8*j
                s = (init+4)/s_max
                mi[j] = np.array([s**3,s**2,s,1])
                for l in range(init,init+8) :
                    x,y = midline[l]
                    d = int(k*(y-horizon)/2.)
                    if x+d >= width :
                        d = width-x-2
                    if x-d < 0 :
                        d = x
                    if y < 0 :
                        y = 0 
                    if y >= height :
                        y = height-1
                    eix[j] += dx[y][x+d] + dx[y][x-d]
                    eiy[j] += dy[y][x+d] + dy[y][x-d]
                    ek += dx[y][x-d] - dx[y][x+d]
            mi = np.dot(mi,C)
            mi = np.dot(np.linalg.inv(np.dot(mi.T,mi)), mi.T) # 4x5
            dqx = np.dot(mi, eix) #4x1
            dqy = np.dot(mi,eiy)
            sQx[ii:ii+4] = Qx[ii:ii+4] + dqx*gamma
            sQy[ii:ii+4] = Qy[ii:ii+4] + dqy*gamma
        dk = ek*0.5e-2
        midline = np.rint(bspline(np.column_stack((sQx,sQy)))).astype(int)
        if (dk > 0.01 or np.sum(np.abs(sQx-Qx)+np.abs(sQy-Qy)) > 17):
            done = False
        else :
            k += dk
            Qx = sQx
            Qy = sQy
```
